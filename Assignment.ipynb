Question 1:
Create a function with one argument formed in DataFrame to check the
data type, the number of null values, the percentage of null values and
the number of unique values for each column!

```python
def check_data(df):
  summary = pd.DataFrame({
      'Data type':df.dtypes,
      'null value':df.isnull().sum(),
      'percetage_null_value':(df.isnull()*100).sum(),
      'unique_value':df.nunique()
  })
  return summary
```

```python
check_data(df)
```

Question 5:
• Show the number of reservations per month of arrival for each
type of hotel!
• Then in which month there are the most reservations in each type
of hotel? Make a conclusion whether the trend is the same in both
types of hotels?
• Do as point B but with the name of the month that has been mapped
into months in numbers!

```python
monthly_reservations = df.groupby(['hotel', 'arrival_date_month']).size().reset_index(name='reservations')

print(monthly_reservations)
```

```python
max_month = monthly_reservations.loc[monthly_reservations.groupby('hotel')['reservations'].idxmax()]

print("Most booked month for each hotel:\n")
print(max_month)
```

```python
month_order = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

df['arrival_month_num'] = df['arrival_date_month'].map(month_order)

monthly_by_num = df.groupby(['hotel', 'arrival_month_num']).size().reset_index(name='reservations')

max_month_num = monthly_by_num.loc[monthly_by_num.groupby('hotel')['reservations'].idxmax()]

print("Most booked (numeric month) for each hotel:\n")
print(max_month_num)
```

Question 7:
Create two dataframe containing:
• Daily reservation (df_daily_reservation)
• Average of daily reservation per
week
(df_avg_daily_reservation_per_week)

```python
df_daily_reservation = df.groupby('arrival_date').size().reset_index(name='daily_reservations')

df_daily_reservation.head()
```

```python
df_daily_reservation['weekday'] = df_daily_reservation['arrival_date'].dt.dayofweek

df_daily_reservation['weekday_name'] = df_daily_reservation['arrival_date'].dt.day_name()
```

```python

df_avg_daily_reservation_per_week = df_daily_reservation.groupby('weekday_name')['daily_reservations'].mean().round(2).reset_index()

days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_avg_daily_reservation_per_week['weekday_name'] = pd.Categorical(df_avg_daily_reservation_per_week['weekday_name'], categories=days_order, ordered=True)
df_avg_daily_reservation_per_week = df_avg_daily_reservation_per_week.sort_values('weekday_name')

df_avg_daily_reservation_per_week
```

Question 9:
• How many average guests stay for each reservation?
• Based on the dataset, what is the highest number of guests? Also
show the reservation data row that has the highest number of guests.

```python
df[['children', 'babies']] = df[['children', 'babies']].fillna(0)

df['total_guests'] = df['adults'] + df['children'] + df['babies']

avg_guests = df['total_guests'].mean().round(2)

print(f"🔹 Average number of guests per reservation: {avg_guests}")
```

```python
max_guests = df['total_guests'].max()
print(f"🔹 Highest number of guests in a reservation: {max_guests}")

df_max_guests = df[df['total_guests'] == max_guests]

df_max_guests
```
